```python
import matplotlib.pyplot as plt
```

```python
f = lambda x: (x + 3)**2
df = lambda x: 2 * (x + 3)
```

```python
x = 2                # Starting point
lr = 0.01            # Learning rate
precision = 1e-6     # Stopping condition
previous_step_size = 1
max_iter = 10000
iters = 0
```

```python
x_list = [x]
y_list = [f(x)]
```

```python
while previous_step_size > precision and iters < max_iter:
    prev_x = x
    x = x - lr * df(prev_x)
    previous_step_size = abs(x - prev_x)
    iters += 1
    x_list.append(x)
    y_list.append(f(x))
    print(f"Iteration {iters}: x = {x:.6f}, f(x) = {f(x):.6f}")
```

```python
print(f"\nLocal minimum occurs at x = {x:.6f}")
print(f"Minimum value of function = {f(x):.6f}")
```

```python
plt.plot(x_list, y_list, 'o-', label='Gradient descent path')  # removed 'r' for default color
plt.title("Gradient Descent on y = (x + 3)^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
