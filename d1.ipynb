```python
# Recursive Fibonacci function
def fib_recursive(n):
    if n <= 1:  # Base case
        return n
    else:  # Recursive call
        return fib_recursive(n - 1) + fib_recursive(n - 2)

# Non-recursive (Iterative) Fibonacci function
def fib_iterative(n):
    if n <= 1:  # Base case
        return n
    a, b = 0, 1  # First two Fibonacci numbers
    for _ in range(2, n + 1):  # Loop to calculate remaining
        a, b = b, a + b
    return b

# Take input from user
n = int(input("Enter the position (n) for Fibonacci: "))

# Output results
print("\nFibonacci number at position", n, "using recursion:", fib_recursive(n))
print("Fibonacci number at position", n, "using iteration:", fib_iterative(n))

# Display full series
print("\nFibonacci Series up to", n, "terms (iterative):")
for i in range(n):
    print(fib_iterative(i), end=" ")

# ----------------------------
# Complexity Analysis
# Recursive: Time = O(2^n), Space = O(n)
# Iterative: Time = O(n), Space = O(1)
```
