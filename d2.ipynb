```python
# Huffman Encoding using Greedy Strategy

# function to find Huffman codes
def huffman(characters, frequency):
    n = len(characters)
    print("\nStep-by-step Huffman Encoding Process:")
    while len(frequency) > 1:
        # find two smallest frequencies
        first_min = min(frequency)
        frequency.remove(first_min)
        second_min = min(frequency)
        frequency.remove(second_min)

        # combine them
        new_node = first_min + second_min
        frequency.append(new_node)

        print(f"Combine nodes ({first_min}, {second_min}) -> New node: {new_node}")

    print("\nFinal Huffman Tree Frequency:", frequency[0])

# main program
n = int(input("Enter number of characters: "))
characters = []
frequency = []

for i in range(n):
    ch = input(f"Enter character {i+1}: ")
    freq = int(input(f"Enter frequency of {ch}: "))
    characters.append(ch)
    frequency.append(freq)

huffman(characters, frequency)
```
