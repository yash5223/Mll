```python
# 0-1 Knapsack Problem using Dynamic Programming

def knapsack(weights, profits, capacity):
    n = len(profits)
    # create DP table of size (n+1) x (capacity+1)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    # build table bottom-up
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                # include item or exclude item
                dp[i][w] = max(profits[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]

    # final cell has the answer
    return dp[n][capacity]

# main program
n = int(input("Enter number of items: "))
weights = []
profits = []

for i in range(n):
    w = int(input(f"Enter weight of item {i+1}: "))
    p = int(input(f"Enter profit of item {i+1}: "))
    weights.append(w)
    profits.append(p)

capacity = int(input("Enter capacity of knapsack: "))

max_profit = knapsack(weights, profits, capacity)
print("\nMaximum Profit:", max_profit)
```
